```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("benchmark.csv")
data
```

```python
plt.plot(np.log(data["naive"]), label="Naive")
plt.plot(np.log(data["tiled"]), label="Tiled")

plt.xticks(range(6), labels=data["size"])
yl = np.linspace(-4, 4, 9)
plt.yticks(yl, labels=[f"{np.exp(l):.3f}" for l in yl])
plt.xlabel("Matrix size")
plt.ylabel("Time [s]")

plt.legend()
plt.show()
```
